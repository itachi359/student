```python
import pandas as pd

# Load the customer reviews dataset
df = pd.read_csv(r"C:\Users\PALADUGULLA KARTHIK\Desktop\Tours_and_Travels.csv")

# Display the first few rows of the dataset
print(df.head())

# Get a summary of the dataset
print(df.info())
# Identify missing values in Review_Text, Rating, and Customer_Age
missing_values = df.isna().sum()
print("Missing values before imputation:\n", missing_values)

# Impute missing numerical values (e.g., Customer_Age)
from sklearn.impute import SimpleImputer

age_imputer = SimpleImputer(strategy='mean')
df['Customer_Age'] = age_imputer.fit_transform(df[['Customer_Age']])

# For missing textual data in Review_Text, use NLP techniques
# Example: Fill missing reviews with a placeholder or use a model to generate text
df['Review_Text'].fillna('No review provided', inplace=True)

# Check for missing values after imputation
missing_values_after = df.isna().sum()
print("Missing values after imputation:\n", missing_values_after)
# Identify duplicate records
duplicates = df.duplicated().sum()
print(f"Number of duplicate records: {duplicates}")

# Remove duplicates
df = df.drop_duplicates()

# Verify that duplicates have been removed
duplicates_after = df.duplicated().sum()
print(f"Number of duplicate records after removal: {duplicates_after}")
# Standardize Rating values (ensure they range between 1-5)
df['Rating'] = df['Rating'].clip(lower=1, upper=5)

# Correct spelling inconsistencies in Tour_Package names
# Example: Replace common misspellings
df['Tour_Package'] = df['Tour_Package'].str.replace('Tropical Paradise', 'Tropical Paradise', case=False)
df['Tour_Package'] = df['Tour_Package'].str.replace('Mountain Adventure', 'Mountain Adventure', case=False)
# Add more replacements as necessary
import seaborn as sns
import matplotlib.pyplot as plt

# Use boxplots to identify outliers in Package_Price and Rating
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Package_Price'])
plt.title('Boxplot of Package Price')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Rating'])
plt.title('Boxplot of Rating')
plt.show()

# Apply transformation techniques if necessary
# Example: Capping outliers for Package_Price
Q1 = df['Package_Price'].quantile(0.25)
Q3 = df['Package_Price'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping the outliers
df['Package_Price'] = df['Package_Price'].clip(lower=lower_bound, upper=upper_bound)
# Convert categorical data into numerical format where required
df['Tour_Package'] = df['Tour_Package'].astype('category').cat.codes

# Check the final structure of the DataFrame
print(df.info())
# Save the cleaned dataset as a CSV file for further analysis
df.to_csv('cleaned_travel_customer_reviews.csv', index=False)
print("Cleaned dataset saved as 'cleaned_travel_customer_reviews.csv'.")
```
