```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("C:/Users/PALADUGULLA KARTHIK/Desktop/dataset1.csv")  # Replace 'dataset.csv' with your file path

# Data Cleaning
# Check for missing values
print("Missing Values:\n", data.isnull().sum())

# Drop rows with missing values (if needed)
data = data.dropna()

# Check for duplicates
print("Duplicates:", data.duplicated().sum())
data = data.drop_duplicates()

# Visualization 1: Bar Plot - Average Effectiveness per Drug by Region
avg_effectiveness = data.groupby(['Product', 'Region'])['Effectiveness'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Effectiveness', hue='Product', data=avg_effectiveness, palette='tab10')
plt.title('Average Effectiveness per Drug Across Regions', fontsize=14)
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.xticks(rotation=45)
plt.legend(title='Product')
plt.tight_layout()
plt.show()

# Visualization 2: Violin Plot - Distribution of Effectiveness and Side Effects by Product
plt.figure(figsize=(12, 6))
sns.violinplot(x='Product', y='Effectiveness', data=data, palette='muted', inner='quartile')
plt.title('Distribution of Effectiveness by Product', fontsize=14)
plt.xlabel('Product')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Product', y='Side_Effects', data=data, palette='muted', inner='quartile')
plt.title('Distribution of Side Effects by Product', fontsize=14)
plt.xlabel('Product')
plt.ylabel('Side Effects')
plt.tight_layout()
plt.show()

# Visualization 3: Pairplot - Relationships Between Effectiveness, Side Effects, and Marketing Spend
sns.pairplot(data[['Effectiveness', 'Side_Effects', 'Marketing_Spend']], diag_kind='kde', kind='scatter', palette='husl')
plt.suptitle('Pairplot of Effectiveness, Side Effects, and Marketing Spend', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()

# Visualization 4: Boxplot - Effectiveness Across Trial Periods
plt.figure(figsize=(10, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=data, palette='pastel')
plt.title('Effectiveness Across Trial Periods', fontsize=14)
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# Visualization 5: Regression Plot - Marketing Spend vs Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=data, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Regression Plot: Marketing Spend vs Effectiveness', fontsize=14)
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# Analysis
print("\nAnalysis:")
print("1. The bar plot reveals which drug performs better in different regions.")
print("2. The violin plots show how effectiveness and side effects vary for each product.")
print("3. The pairplot highlights relationships between effectiveness, side effects, and marketing spend.")
print("4. The boxplot provides insight into how drug effectiveness changes over different trial periods.")
print("5. The regression plot illustrates the influence of marketing spend on drug effectiveness.")
```
