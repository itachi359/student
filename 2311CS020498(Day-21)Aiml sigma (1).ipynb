```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Civil_Engineering_Regression_Dataset.csv')


print("First 5 rows of the dataset:")
print(df.head())


print("\nColumn Names:")
print(df.columns)


dependent_var = 'Construction_Cost'  
if dependent_var in df.columns:
    
    independent_vars = df.columns.drop([dependent_var, 'Project_ID'])  
else:
    print(f"Error: '{dependent_var}' not found in dataset.")
    independent_vars = df.columns  

print("\nIndependent Variables (Features):", independent_vars.tolist())
print("Dependent Variable (Target):", dependent_var)

missing_values = df.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values[missing_values > 0])


df.fillna(df.mean(), inplace=True) 
df.fillna(df.mode().iloc[0], inplace=True)  


print("\nSummary Statistics:")
print(df.describe().T[['mean', '50%', 'std']])  


plt.figure(figsize=(10, 6))
correlation_matrix = df.corr()  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Features")
plt.show()
```
