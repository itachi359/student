```python
import pandas as pd

# Load the healthcare dataset
df = pd.read_csv(r"C:\Users\PALADUGULLA KARTHIK\Desktop\Healthcare_Data.csv")

# Display the first few rows of the dataset
print(df.head())

# Get a summary of the dataset
print(df.info())
# Identify missing values in patient demographics
missing_values = df.isna().sum()
print("Missing values before imputation:\n", missing_values)

# Apply appropriate imputation methods
# For numerical columns (e.g., age, blood pressure)
from sklearn.impute import SimpleImputer

# Mean imputation for age and blood pressure
age_imputer = SimpleImputer(strategy='mean')
df['age'] = age_imputer.fit_transform(df[['age']])

bp_imputer = SimpleImputer(strategy='mean')
df['blood_pressure'] = bp_imputer.fit_transform(df[['blood_pressure']])

# For categorical columns (e.g., gender)
gender_imputer = SimpleImputer(strategy='most_frequent')
df['gender'] = gender_imputer.fit_transform(df[['gender']])

# Check for missing values after imputation
missing_values_after = df.isna().sum()
print("Missing values after imputation:\n", missing_values_after)
# Identify duplicate records
duplicates = df.duplicated().sum()
print(f"Number of duplicate records: {duplicates}")

# Remove duplicates
df = df.drop_duplicates()

# Verify that duplicates have been removed
duplicates_after = df.duplicated().sum()
print(f"Number of duplicate records after removal: {duplicates_after}")
import seaborn as sns
import matplotlib.pyplot as plt

# Use boxplots to identify outliers in numerical columns
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['age'])
plt.title('Boxplot of Age')
plt.show()

# Apply transformations or capping techniques to handle outliers
# Example: Capping outliers for age
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping the outliers
df['age'] = df['age'].clip(lower=lower_bound, upper=upper_bound)
import seaborn as sns
import matplotlib.pyplot as plt

# Use boxplots to identify outliers in numerical columns
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['age'])
plt.title('Boxplot of Age')
plt.show()

# Apply transformations or capping techniques to handle outliers
# Example: Capping outliers for age
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping the outliers
df['age'] = df['age'].clip(lower=lower_bound, upper=upper_bound)
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Convert categorical variables into numerical representations
label_encoder = LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender'])

# Scale numerical variables using Min-Max Scaling
scaler = MinMaxScaler()
df[['age', 'blood_pressure']] = scaler.fit_transform(df[['age', 'blood_pressure']])

# Alternatively, you can use Standard Scaling
# standard_scaler = StandardScaler()
# df[['age', 'blood_pressure']] = standard_scaler.fit_transform(df[['age', 'blood_pressure']])
# Ensure no missing values or duplicates remain
assert df.isna().sum().sum() == 0, "There are still missing values in the dataset."
assert df.duplicated().sum() == 0, "There are still duplicate records in the dataset."

# Check data types and correct inconsistencies
print(df.dtypes)

# If necessary, convert data types
# Example: df['age'] = df['age'].astype(int)
# Save the cleaned dataset as a CSV file for further analysis
df.to_csv('cleaned_healthcare_dataset.csv', index=False)
print("Cleaned dataset saved as 'cleaned_healthcare_dataset.csv'.")
```
