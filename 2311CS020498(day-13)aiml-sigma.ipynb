```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("C:/Users/PALADUGULLA KARTHIK/Desktop/dataset.csv")  # Replace 'dataset.csv' with your file path

# Data Cleaning
# Check for missing values
print("Missing Values:\n", data.isnull().sum())

# Drop rows with missing values (if needed)
data = data.dropna()

# Check for duplicates
print("Duplicates:", data.duplicated().sum())
data = data.drop_duplicates()

# Visualization 1: Bar Plot - Total Sales Per Region
sales_per_region = data.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=sales_per_region, palette='viridis')
plt.title('Total Sales per Region', fontsize=14)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 2: Scatter Plot - Marketing Spend vs Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=data, hue='Region', palette='cool')
plt.title('Marketing Spend vs Sales', fontsize=14)
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# Visualization 3: Boxplot - Drug Effectiveness Across Age Groups
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age_Group', y='Effectiveness', data=data, palette='Set2')
plt.title('Drug Effectiveness Across Age Groups', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# Visualization 4: Line Plot - Sales Trend Over Trial Periods
sales_trend = data.groupby(['Product', 'Trial_Period'])['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='Trial_Period', y='Sales', hue='Product', data=sales_trend, marker='o')
plt.title('Sales Trend Over Trial Periods', fontsize=14)
plt.xlabel('Trial Period')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# Visualization 5: Heatmap - Correlation Between Sales, Marketing Spend, and Effectiveness
correlation_matrix = data[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()

# Summary of Observations
print("\nSummary of Observations:")
print("- Analyze the scatter plot to observe the relationship between marketing spend and sales.")
print("- Use the boxplot to determine which age groups have higher drug effectiveness.")
print("- The bar plot indicates which regions have the highest and lowest sales.")
print("- The line plot reveals sales trends across trial periods for each product.")
print("- The heatmap shows correlations between sales, marketing spend, and effectiveness.")
```
